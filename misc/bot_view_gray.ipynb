```python
%matplotlib inline
import matplotlib.pyplot as plt
from skimage import color, io, transform
import glob
```

```python
#SIZE = (256, 256)
imgs = [transform.resize(io.imread(img), SIZE) for img in glob.glob('./images/*.bmp')]
```

```python
cnt = 0
plt.figure(num=None, figsize=(12, 20))
for n, im in enumerate(imgs):
    plot_im = [('Original', im), 
               ('Red', im[:,:,0]), 
               ('Green', im[:,:,1]), 
               ('Blue', im[:,:,2]), 
               ('Gray', color.rgb2gray(im)), 
               ('Value of HSV', color.rgb2hsv(im)[:,:,2]), 
               ('Luminosity', color.rgb2luv(im)[:,:,0])]
    for s, i in plot_im:
        cnt+=1
        plt.subplot(len(imgs), len(plot_im), cnt)
        if n == 0:
            plt.title(s, fontsize=15)
        if cnt%len(plot_im)==0:
            plt.gray()
        plt.axis('off')
        plt.grid('off')
        plt.tick_params(labelbottom='off', labelleft='off')
        plt.imshow(i, interpolation='nearest', aspect=1)
plt.tight_layout()
```
